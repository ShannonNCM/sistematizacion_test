```python
# se importan las librerias a utilizar
import spacy
import pandas as pd
from collections import Counter
```

```python
# se carga el modelo NLP en espanol
nlp = spacy.load('es_core_news_md')
```

```python
# se importa la informacion del archivo de excel
df = pd.read_excel('info.xlsx', sheet_name='Hoja1')

# se divide la informacion en distintos dataframes de acuerdo a la pregunta de la que proviene (agregue un
# drop_na porque hay algunas que estaban vacias)
ans_p1 = df[['Sector','P1']].dropna()
ans_p2 = df[['Sector','P2']].dropna()
ans_p3 = df[['Sector','P3']].dropna()
```

```python
# se define la funcion que se va a usar para poder extraer las palabras clave
def extract_keywords(text, dict):
    '''esta parte aplica el modelo de lenguaje spacy al texto que se esta ingresando y se guarda en 'doc'
    'doc' es un objeto Doc de spacy, que es un contendor para el texto procesado, contiene 'tokens'
    individuales (palabras o puntuaciones) junto con toda la informacion linguistica que spacy ha 
    computado'''
    doc = nlp(text)
    '''se usa 'set' para evitar duplicados, por ejemplo, si un mismo termino aparece mas de una vez entonces
    solamente se va a guardar una vez'''
    extracted_concepts = set()
    
    '''se hace un loop en doc por cada palabra (token) del texto que se proceso'''
    for token in doc:
        '''se obtiene la forma base (lemma) del token'''
        lemma = token.lemma_
        '''se revisa si el lemma coincide con alguna de las palabras en los terminos del diccionario
        '''
        for termino, keywords in dict.items():
            if lemma in keywords:
                extracted_concepts.add(termino)
    return list(extracted_concepts)
```

```python
# se define un diccionario con las palabras clave que se quieren buscar dentro del texto que se esta analizando
dict = {
    'Datos' : ['dato'],
    'Agenda digital' : ['cambio de paradigma'],
    'Capacidades digitales' : ['capacitación', 'alafabetización', 'educación', 'formación', 'capacitar', 'educativo'],
    'Gobernanza' : ['gobernanza', 'gobierno'],
    'Brecha digital' : ['acceso a todos', 'brecha digital', 'implementar programas'],
    'Marco legal' : ['ley', 'legal', 'normar', 'legalizar'],
    'Infraestructura publica' : ['infraestructura', 'acceso a electricidad', 'equipo tecnológico', 'acceso a internet', 'internet a nivel nacional'],
    'Digitalizacion de la informacion' : ['digitalizar', 'sistema'],
    'Participacion ciudadana' : ['libertad de expresión']}
```

```python
# se aplica la funcion para obtener los datos que queremos
ans_p1['res'] = ans_p1['P1'].apply(lambda x: extract_keywords(x, dict))
ans_p1
```
